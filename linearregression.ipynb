```python
import json
from pandas.io.json import json_normalize
```

```python
def categorical_genres(df):
    list = df['genres']
    df_temp = []
    features_genres = ['Action', 'Adventure', 'Fantasy', 'Science Fiction' ]
    for i in list:
        json_data = json.loads(i)
        movie_genres = []
        for x in range(len(json_data)):
            if (json_data[x]['name'] in features_genres):
                movie_genres.append(json_data[x]['name'])
            elif (json_data[x]['name'] not in features_genres) and (json_data[x]['name'] != ''):
                movie_genres.append("Other")    
        
        df_temp.append(','.join(movie_genres))
    
    df_new_genres = pd.DataFrame({'All new genres':df_temp})   
    return df_new_genres
```

```python
df_genres = categorical_genres(df)
df = pd.concat([df, df_genres['All new genres'].str.get_dummies(sep=',')], axis=1)
```

```python
df_numeric = df._get_numeric_data()
```

```python
# 4. Split dataset into training and test datasets in 0.7/0.3
# 5. Select all numerical features that can be used to predict ‘vote average’ target

A = df_numeric[['popularity','revenue','runtime','vote_count','Action','Adventure','Fantasy','Other','Science Fiction']]
B = df_numeric['vote_average']
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(A, B, train_size=0.7, random_state=0)
```

```python
# 9. Use score to identify most accurate and predictive features. Justify the final choice of the features.

regressor = LinearRegression()  
regressor.fit(Xtrain, Ytrain)
regressor.score(Xtest, Ytest)
```
